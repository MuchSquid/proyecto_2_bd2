```python
import requests
import re
import librosa
import os
```

```python
def getLibrosaFeatures(spotify_df):
    librosa_features = [f"MFCC{it+1}" for it in range(50)]
    librosa_data = []  # Lista para almacenar las filas de datos

    folder_path = 'mp3'  # Carpeta donde están almacenados los MP3

    for i in range(len(spotify_df)):
        # Obtener el track_id y construir el path del MP3
        track_id = spotify_df.iloc[i].track_id
        file_path = os.path.join(folder_path, f"{track_id}.mp3")
        
        # Inicializar fila con la ruta del MP3
        row_data = {f"MFCC{cc+1}": None for cc in range(50)}
        row_data['mp3'] = file_path  # Añadir la ruta como columna mp3

        # Verificar si el archivo existe y procesarlo
        if os.path.exists(file_path):
            try:
                # Cargar archivo y calcular MFCCs
                audio_time_series, sampling_rate = librosa.load(file_path)
                mfcc_array = librosa.feature.mfcc(y=audio_time_series, sr=sampling_rate, n_mfcc=50)
                mean_mfcc_array = np.mean(mfcc_array, axis=1)  # Promedio de MFCCs
                
                # Guardar MFCCs en la fila
                for cc in range(50):
                    row_data[f"MFCC{cc+1}"] = mean_mfcc_array[cc]
            except Exception as e:
                print(f"Error procesando {file_path}: {e}")
        else:
            print(f"Archivo MP3 no encontrado: {file_path}")
        
        # Agregar la fila a la lista
        librosa_data.append(row_data)

    # Crear DataFrame a partir de las filas
    librosa_df = pd.DataFrame(librosa_data)
    return librosa_df
librosa_features_df = getLibrosaFeatures(music)
```

```python
from tqdm import tqdm
from concurrent.futures import ThreadPoolExecutor
```

```python
def process_file(file_info):
    track_id, folder_path = file_info
    file_path = os.path.join(folder_path, f"{track_id}.mp3")
    row_data = {f"MFCC{cc+1}": None for cc in range(50)}
    row_data['mp3'] = file_path  # Ruta del archivo MP3

    if os.path.exists(file_path):
        try:
            # Cargar archivo con tasa de muestreo fija
            audio_time_series, sampling_rate = librosa.load(file_path, sr=22050)
            mfcc_array = librosa.feature.mfcc(y=audio_time_series, sr=sampling_rate, n_mfcc=50)
            mean_mfcc_array = np.mean(mfcc_array, axis=1)  # Promedio de MFCCs
            
            # Guardar MFCCs en la fila
            for cc in range(50):
                row_data[f"MFCC{cc+1}"] = mean_mfcc_array[cc]
        except Exception as e:
            print(f"Error procesando {file_path}: {e}")
    else:
        print(f"Archivo MP3 no encontrado: {file_path}")
    
    return row_data
```

```python
def getLibrosaFeatures(spotify_df):
    folder_path = 'mp3'  # Carpeta donde están almacenados los MP3
    file_info_list = [(spotify_df.iloc[i].track_id, folder_path) for i in range(len(spotify_df))]

    # Usar ThreadPoolExecutor para procesar en paralelo
    with ThreadPoolExecutor(max_workers=8) as executor:
        # tqdm para mostrar el progreso
        librosa_data = list(tqdm(executor.map(process_file, file_info_list), total=len(file_info_list)))

    # Crear DataFrame a partir de los resultados
    librosa_df = pd.DataFrame(librosa_data)
    return librosa_df

# Llamar a la función con tu DataFrame de Spotify
librosa_features_df = getLibrosaFeatures(music)
```

```python
columnas_mfcc = [f"MFCC{i}" for i in range(1, 51)]
librosa_features_df['MFCC_Vector'] = librosa_features_df[columnas_mfcc].apply(lambda row: row.tolist(), axis=1)

# Eliminar las columnas de MFCC individuales
librosa_features_df.drop(columnas_mfcc, axis=1, inplace=True)
```

```python
librosa_features_df['MFCC_Vector'] = librosa_features_df['MFCC_Vector'].apply(np.array)
```
